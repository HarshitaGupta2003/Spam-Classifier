## DATA PREPROCESSING
##lower case
##tokenization
##remove special characters
##remove stopwords and punctuations
##stemming

```python
# for this we will crete a function that will perform all these functionality at once
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
string.punctuation
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps.stem('loving')
```

```python
def text_transform(text):
    #change in lower case
    text=text.lower()
    #Tokenize or split 
    text=nltk.word_tokenize(text)
    y = []
    #Removing special characters
    for i in text:
        if i.isalnum():
         y.append(i)
    # clear y list to use it again
    text=y[:]
    y.clear()
    # remove stopwords and punctuation
    for i in text :
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text=y[:]
    y.clear()
    #Stemming
    for i in text :
      y.append(ps.stem(i))
    
    return " ".join(y)
    
```

```python
df['Text'][3]
```

```python
text_transform('U dun say so early hor... U c already then say...')
```

```python
df['Transformed_text']=df['Text'].apply(text_transform)
```

##  MODEL BUILDING 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
x = cv.fit_transform(df['Transformed_text']).toarray()
```

```python
y=df['Target'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
#objext creation
gnb=GaussianNB()
mnb=MultinomialNB()
bnb=BernoulliNB()
```

```python
#traning  gnb model using trainig data
gnb.fit(x_train,y_train)
```

```python
#prediction on input testing data(x_test)
y_pred1=gnb.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
bnb.fit(x_train,y_train)
y_pred2=bnb.predict(x_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

```python
mnb.fit(x_train,y_train)
y_pred3=mnb.predict(x_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
```
